## Outliers Removal

#### Using Z-score Method:

```python
from scipy import stats
from scipy.stats import zscore
import warnings
warnings.filterwarnings('ignore')
```

<p> Threshold: Threshold is number of std away from mean at which the data point is considered as an outlier </p>
    <p> Standard Threshold: -3 or 3 </p>
    <p> This means any data point less than -3 or greater than 3 will be considered as an outlier. </p>
        <p> In our case, threshold = 3 </p>
    

```python
def remove_outliers_using_z_score(data_columns,threshold=3):
    z_scores = zscore(data_columns)
    filtered_data = data_columns[np.abs(z_scores)<threshold]
    return filtered_data
```

```python
for column in df1:
    df1[column] = remove_outliers_using_z_score(df1[column])
```

## HyperParameter Tuning:

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [0.1, 1, 10, 100],             
    'gamma': [1, 0.1, 0.01, 0.001],      
    'kernel': ['rbf', 'poly', 'sigmoid'] 
}
```

```python
# Initializing SVC
svm = SVC()
```

```python
# Perform Grid Search with cross-validation
grid_search = GridSearchCV(estimator=svm, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)
```

```python
# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

```python
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

```python
# Predict on the test data using the best found model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
```

#### Evaluate the model:

```python
print("Test Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
```
